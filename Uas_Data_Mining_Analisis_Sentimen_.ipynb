## Preprocessing

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Menghapus URL
df['content'] = df['content'].str.replace('https\S+', ' ', case=False, regex=True)
```

```python
# Huruf Kecil
df['content'] = df['content'].str.lower()
```

```python
# Menghapus mention
df['content'] = df['content'].str.replace('@\S+', ' ', case=False, regex=True)
```

```python
# Menghapus Hashtag
df['content'] = df['content'].str.replace('#\S+', ' ', case=False, regex=True)
```

```python
# Menghapus Next karakter
df['content'] = df['content'].str.replace("\'\w+", ' ', case=False, regex=True)
```

```python
# Menghapus Tanda Baca
df['content'] = df['content'].str.replace("[^\w\s]", ' ', case=False, regex=True)
```

```python
# Menghapus Extra Whitespace
df['content'] = df['content'].str.replace("\s(2)", ' ', case=False, regex=True)
```

**Tokenizing**

```python
# impor tokenize dari modul nltk
from nltk.tokenize import word_tokenize
```

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer('\w+')
df['content_token']=df['content'].apply(regexp.tokenize)
df.head(3)
```

**Filtering**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

# daftar stopwords
stopwords = nltk.corpus.stopwords.words("indonesian")
my_stopwords = ['livinbymandiri']
stopwords.extend(my_stopwords)
```

```python
# hapus stopwords
df['content_token'] = df['content_token'].apply(lambda x: [item for item in x if item not in stopwords])
df.head(3)
```

## Stemming sastrawi

```python
!pip install Sastrawi
```

```python
# import Sastrawi package
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
df['stemmed'] = df['content_token'].apply(lambda x: [stemmer.stem(y) for y in x])
```

```python
df.head(5)
```

## Hanya yang lebih dari 3 kata yang diambil

```python
df['text_string'] = df['stemmed'].apply(lambda x: ' '.join([item for item in x if len(item)>3]))
```
